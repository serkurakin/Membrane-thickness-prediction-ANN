# Предсказание толщины мембраны из текстового файла с данными малоуглового рассеяния

```python
import tensorflow as tf
import numpy as np
import joblib
import pandas as pd
```

```python
model = tf.keras.models.load_model('thickness_model.keras')
scaler = joblib.load('scaler.pkl')
```

```python
def predict(model, scaler, filename):


    data = np.loadtxt(filename, skiprows=1)
    q_orig, I_orig = data[:, 0], data[:, 1]


    q_range_train = np.logspace(-2.06, -0.34, 100)
    I_interp = np.interp(q_range_train, q_orig, I_orig)

    if I_interp[0] > 0:
        I_interp = I_interp * (300 / I_interp[0])

    I_processed = I_interp.reshape(1, -1)

    I_normalized = scaler.transform(I_processed)


    prediction = model.predict(I_normalized, verbose=0)
    thickness = prediction[0][0] * 1.077
    print(f"толщина равна: {thickness:.2f} A")

    return thickness
```

```python
result = predict(model, scaler, '/content/DPPC_DOPC_50_50_ULV_t20.dat')
```
